```python
import copy

def calculate_magnitude(x_db,lr_dim):
    cent = x_db[3*(lr_dim//2):5*(lr_dim//2), 3*(lr_dim//2):5*(lr_dim//2)]
    cent_pwr = cent.sum()
    
    temp = copy.deepcopy(x_db)
    temp[lr_dim:3*lr_dim, lr_dim:3*lr_dim] = 0
    rim_pwr = temp.sum()
    return cent_pwr,rim_pwr
```
